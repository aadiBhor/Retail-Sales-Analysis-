# 1) Demographic Influence on Purchasing Behavior 

```python
plt.figure(figsize=(7, 5))
sns.barplot(data=count_df,x='Age group',y='Count',hue='Product Category',errorbar=('ci', False),palette="Set2")
plt.title('Age Group Influence on Product Purchases')
plt.xlabel("Age Group")
plt.ylabel("Number of Purchases")
plt.legend(title='Product Category')
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

### Adults are the main buyers across all categories, especially Clothing and Electronics, while Seniors buy moderately and Teens contribute the least.

```python
plt.figure(figsize=(7, 6))
sns.barplot(data=count_df,x='Product Category',y='Count',hue='Gender',errorbar=('ci', False),palette="Set1")
plt.ylim(0,80)
plt.title("Product Category Preferences by Gender")
plt.ylabel("Number of Purchases")
plt.legend()
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

### Females buy more Beauty products, while Clothing and Electronics purchases are almost equal between males and females.

## Adults are the main buyers across all categories, while females favor Beauty products and both genders equally purchase Clothing and Electronics.

# 2) Sales Patterns Across Time Periods

```python
monthly_sales=df.groupby(['year','month'])['Total Amount'].sum().reset_index()
monthly_sales
```

```python
yearly_sales = df.groupby("year")['Total Amount'].sum().reset_index()

sns.barplot(data=yearly_sales, x="year", y="Total Amount")
plt.title("Yearly Sales Trend")
plt.show()
```


### The Sales is Very High in 2023 and Total Revanue is also High, while 2024 data is still incomplete so its show Low Sales and Low Total Revanue

# 4) Customer Spending and Product Preferences 
Study correlations between age, total spending, and product choices.    
Identify high-value customer segments based on spending behavior.

```python
sns.scatterplot(data=df,x='Age',y='Total Amount',hue='Product Category')
plt.title("Age vs Spending by Product Category")
plt.ylabel('Spending')
plt.show()
```

### Teens and Adults are the most valuable customer groups, while Seniors represent a smaller spending segment.

```python
customer_spend = df.groupby("Customer ID")["Total Amount"].sum().reset_index()
customer_spend['customer segments'] = customer_spend['Total Amount'].apply(lambda x: "Low" if x <=700 else "Mid" if x <=1400 else "High")
customer_spend
```

```python
df['customer segments']=customer_spend['customer segments']
```

```python
customer_segments_value=customer_spend.groupby("customer segments")["Total Amount"].agg(["count","sum"]).reset_index()
customer_segments_value
```

```python
sns.barplot(data=customer_segments_value, x='customer segments', y='sum', palette="Set2")
plt.title("Revenue Contribution by Customer Segments")
plt.ylabel("Total Revenue")
plt.show()
```

### High-value customers belong to the “High” segment, as they contribute the largest share of revenue despite being fewer in number compared to other segments.

Compare purchasing behavior during different seasons.       
Identify changes in product preferences over time. 

```python
df['month_name']=df['Date'].dt.month_name()
```

```python
df['month_name'].unique()
```

## Spring is the best-performing season, while Winter and Late Winter are weak sales periods, suggesting strong seasonality in demand.

# 7) Product Price Distribution Insights 
 Evaluate the price distribution within each product category.  
 Identify price-sensitive and premium product segments. 

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Product Category", y="Price per Unit")
plt.title("Price Distribution within Each Product Category")
plt.show()
```

### Beauty, Clothing, and Electronics all show similar average prices

```python
df['Price Segment']=df['Price per Unit'].apply(lambda x: "Price-Sensitive" if x <=100 else "Premium")
```

```python
df
```

```python
price_seg=df.groupby(['Product Category','Price Segment'])['Total Amount'].sum().reset_index()
price_seg
```

```python
sns.barplot(data=price_seg,x='Product Category',y='Total Amount',hue='Price Segment',alpha=0.6)
plt.title("Price-Sensitive vs Premium Segments by Product Category")
plt.ylabel("Total Spending")
plt.tight_layout()
plt.grid(alpha=0.6)
plt.show()
```

### Premium buyers spend way more than price-sensitive ones across Beauty, Clothing, and Electronics
### customers spend highest on Premiun product -> Electronics / Customer spend lowest on the Price - sensitive product -> Beauty
